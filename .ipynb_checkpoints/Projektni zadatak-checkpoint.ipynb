```python
# Najprej zdru≈æimo X in y v en sam DataFrame (ƒçe ≈°e nista)
df_full = pd.concat([X_scaled, y], axis=1)

# Ustvarimo binarno tarƒço glede na mediano motor_UPDRS
threshold = df_full['motor_UPDRS'].median()
df_full['motor_UPDRS_bin'] = (df_full['motor_UPDRS'] > threshold).astype(int)

# Preverimo porazdelitev novih vrednosti
df_full['motor_UPDRS_bin'].value_counts()
```

```python
# Funkcija za izraƒçun WoE in IV
def calc_woe_iv(data, feature, target):
    df = data[[feature, target]].copy()
    df['bin'] = pd.qcut(df[feature], q=10, duplicates='drop')
    grouped = df.groupby('bin')[target].agg(['count', 'sum'])
    grouped['good'] = grouped['count'] - grouped['sum']
    grouped['bad'] = grouped['sum']
    
    total_good = grouped['good'].sum()
    total_bad = grouped['bad'].sum()
    
    grouped['%good'] = grouped['good'] / total_good
    grouped['%bad'] = grouped['bad'] / total_bad
    grouped['WoE'] = np.log(grouped['%good'] / grouped['%bad']).replace([np.inf, -np.inf], 0)
    grouped['IV'] = (grouped['%good'] - grouped['%bad']) * grouped['WoE']
    
    return grouped, grouped['IV'].sum()

# Izraƒçun IV za vse atribute
iv_summary = {}
woe_tables = {}

features_to_check = [col for col in X_scaled.columns if col not in ['motor_UPDRS', 'total_UPDRS']]

for col in features_to_check:
    try:
        woe_table, iv = calc_woe_iv(df_full, col, 'motor_UPDRS_bin')
        iv_summary[col] = iv
        woe_tables[col] = woe_table
    except Exception as e:
        print(f"Napaka pri {col}: {e}")

# Prikaz rezultatov
iv_df = pd.DataFrame.from_dict(iv_summary, orient='index', columns=['IV'])
iv_df = iv_df.sort_values(by='IV', ascending=False)

iv_df  # to ti izpi≈°e tabelo IV vrednosti po pomembnosti
```

### WoE and IV Analysis ‚Äì Conclusion

For each variable, we calculated the **Information Value (IV)**, which measures the predictive power of an attribute in distinguishing between patients with milder and more severe forms of Parkinson‚Äôs disease (binary target: `motor_UPDRS_bin`).

- The variable **`age`** (IV = 0.411) proved to be a **very strong predictor**.
- The variable **`Shimmer:APQ11`** (IV = 0.105) also reached the **threshold of moderate predictive power**.
- Several other variables (e.g., **DFA**, **RPDE**, **Shimmer**, **PPE**) showed **moderate predictive strength** and will be included in further modeling steps.
- Variables with **IV < 0.02** (e.g., `test_time`) are **not informative** for classification and will be excluded from the modeling phase.

These results serve as the foundation for variable selection in the next steps, such as **coarse classing**, **predictive model development**, or **Bayesian network conoƒçan), ƒçe ≈æeli≈°.ve mre≈æe.


```python
import numpy as np
import pandas as pd

# 1. Vrni skalirani 'age' nazaj v originalne vrednosti
age_min = 36
age_max = 85
df_full['age_original'] = df_full['age'] * (age_max - age_min) + age_min

# 2. Funkcija za coarse classing glede na realne starosti
def coarse_age_group(age):
    if age <= 58:
        return 'mlaj≈°i (36‚Äì58)'
    elif age <= 72:
        return 'srednji (59‚Äì72)'
    else:
        return 'starej≈°i (73‚Äì85)'

# 3. Ustvari nov stolpec s skupinami
df_full['age_group'] = df_full['age_original'].apply(coarse_age_group)

# 4. Poglej, koliko primerov je v vsaki skupini
print("üì¶ Porazdelitev po skupinah:")
print(df_full['age_group'].value_counts())

# 5. Izraƒçunaj WoE za te nove skupine
woe_grouped = df_full.groupby('age_group')['motor_UPDRS_bin'].agg(['count', 'sum'])
woe_grouped['good'] = woe_grouped['count'] - woe_grouped['sum']
woe_grouped['bad'] = woe_grouped['sum']
woe_grouped['%good'] = woe_grouped['good'] / woe_grouped['good'].sum()
woe_grouped['%bad'] = woe_grouped['bad'] / woe_grouped['bad'].sum()
woe_grouped['WoE'] = np.log(woe_grouped['%good'] / woe_grouped['%bad'])

# 6. Prika≈æi tabelo WoE
print("\nüìä WoE za age_group:")
display(woe_grouped[['count', 'WoE']])
```

### üßæ Storytelling ‚Äì Problem Description and Univariate Analysis

The goal of the analysis was to explore which attributes best predict the **severity of Parkinson‚Äôs disease**, using a binary split of the `motor_UPDRS` variable (less severe = 0, more severe = 1).

We began the process by reviewing all input variables and evaluating their individual ability to predict the target, which is known as **univariate analysis**.

---

#### üìå WoE and IV for Each Attribute

For each attribute, we calculated:
- **Weight of Evidence (WoE)** ‚Äì indicating whether certain values increase or decrease the likelihood of a more severe disease form,
- **Information Value (IV)** ‚Äì measuring the overall predictive power of the attribute.

Based on IV values, the attributes were categorized as follows:
- `age` stood out as a **very strong predictor** (IV = 0.411),
- `Shimmer:APQ11` showed **moderate predictive power** (IV = 0.105),
- Several attributes such as `DFA`, `RPDE`, `Shimmer`, and `PPE` had **medium predictive strength** (IV between 0.05 and 0.1),
- Attributes such as `test_time` (IV = 0.0099) had **no meaningful predictive value** and were excluded from further analysis.

---

#### üìå Selection of Most Predictive Attributes

Based on the IV threshold, we selected only variables with **IV > 0.02**,  
ensuring that only features with at least some predictive power were retained for subsequent modeling (e.g., Bayesian network).

---

#### üìå Coarse Classing ‚Äì Grouping by WoE

To improve interpretability and model stability, we applied **coarse classing** to the `age` attribute, which had the highest predictive power.

Instead of keeping age as a continuous variable, we divided it into three meaningful categories:
- **younger (36‚Äì58)**: WoE = +0.60 ‚Üí lower risk
- **middle-aged (59‚Äì72)**: WoE ‚âà 0 ‚Üí neutral
- **older (73‚Äì85)**: WoE = ‚Äì0.64 ‚Üí higher risk

This grouping allowed for clearer interpretation and better data preparation for the modeling phase.

---

#### üí° Why Did We Choose `age`?

The `age` attribute emerged as the **most predictive variable**, which also aligns with domain knowledge ‚Äì symptoms of Parkinson‚Äôs disease often worsen with age.  
Therefore, it makes sense to include age in the model in an interpretable way through age groupings (coarse classing).

---

This narrative supports all steps of the univariate analysis and sets the foundation for the nexteko ali pripravim ≈°e kratek uvod za poglavje. Povej samo!azo ‚Äì modeliranje z Bayesovo mre≈æo.
